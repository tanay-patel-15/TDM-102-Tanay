```python
import pandas as pd
```

## Question 2

```python
video_games["high_rated"] = video_games["critic_score"] >= 7
```

```python
video_games["high_rated"].value_counts()
```

```python
video_games["low_rated"] = video_games["critic_score"] < 7

video_games["low_rated"].value_counts()
```

```python
video_games['critic_score'].isna().sum() 
```

```python
sports_hits = video_games[
    (video_games["genre"] == "Sports") & (video_games["high_rated"] == True)
]
```

```python
sports_hits[sports_hits['critic_score'] == 9.5]
```

## In sports_hits, Madden NFL 2004, Madden NFL 2005 and SSX 3 had a critic score of exactly 9.5

```python
sports_fails = video_games[
    (video_games["genre"] == "Sports") & (video_games["low_rated"] == True)
]
```

```python
sports_fails[sports_fails['critic_score'] == 2.8]
```

## In sports_fails, Major League Basketball 2K10 and MLB Slam! had a critic score of 2.8

In question 2, we learn how to create a new filtered column from an existing column by putting a limit or filter on the values of the existing column. We created a high rated and a low rated column and then considered that many games have a NA value and we also calculated the number of rows with NA as their critic score and then moving on, we made two new dataframes namely, sports_hits and sports_fails deriving from the dataframe video_games with the filters of genre being sport and high rated & low rated columns being true respectively.

```python
app_store = pd.read_csv("/anvil/projects/tdm/data/grouping/googleplaystore.csv")
```

```python
app_store[["Rating", "Reviews"]].head()
```

```python
app_store.dtypes
```

```python
app_store = app_store[app_store["Clean_Reviews"] != "3.0M"]
```

## The string value "3.0M" was causing an error when creating the numeric reviews column.

```python
# Remove any commas that may exist
# Replace with "" (nothingness)
app_store["Clean_Reviews"] = app_store["Reviews"].str.replace(",", "", regex=False)
# Convert all values to type float (decimal numbers)
# ATTENTION: This line will cause an error message to appear
app_store["Numeric_Reviews"] = app_store["Clean_Reviews"].astype(float)
```

```python
app_store["rating_per_review"] = app_store["Rating"] / app_store["Numeric_Reviews"]
```

```python
app_store[["Rating", "Reviews", "Numeric_Reviews", "rating_per_review"]].head()
```

In question 4, we imported a new Google play Store dataset into a dataframe called app_store and then we played around the data frame and later on, changed all the float values of the reviews column into float type from object type. Then using that numeric reviews column, we created a new rating per review column.
